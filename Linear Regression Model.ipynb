```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
df22 = pd.read_csv("2022.csv")
df21 = pd.read_csv("2021.csv")
df20 = pd.read_csv("2020.csv")
df19 = pd.read_csv("2019.csv")
df18 = pd.read_csv("2018.csv")
df17 = pd.read_csv("2017.csv")
df16 = pd.read_csv("2016.csv")
df15 = pd.read_csv("2015.csv")
```

```python
df19.rename(columns={"Freedom to make life choices":"Freedom",
                     "Score":"Happiness Score",
                     "Country or region":"Country",
                     "Perceptions of corruption":"Trust"
                    }, inplace = True)
df19["Year"]=2019
df19.drop(["Overall rank"], axis=1, inplace=False)
```

# Trend of overall happiness score

```python
scoresummary = df22.describe()[['Happiness Score']]
scoresummary['2015'] = df15.describe()[['Happiness Score']]
scoresummary.drop(columns=['Happiness Score'],axis=1, inplace=True)
scoresummary['2016'] = df16.describe()[['Happiness Score']]
scoresummary['2017'] = df17.describe()[['Happiness Score']]
scoresummary['2018'] = df18.describe()[['Happiness Score']]
scoresummary['2019'] = df19.describe()[['Happiness Score']]
scoresummary['2020'] = df20.describe()[['Happiness Score']]
scoresummary['2021'] = df21.describe()[['Happiness Score']]
scoresummary['2022'] = df22.describe()[['Happiness Score']]
scoresummary.drop('std', inplace=True)
scores = scoresummary.transpose()
scores
```

# Correlation

```python
indep = df22.drop(['RANK','Happiness Score','Country','Whisker-high','Whisker-low',"Year"], axis=1, inplace=False)
plt.figure(figsize=(10,8))
corrMatrix = indep.corr()
sn.heatmap(corrMatrix,annot=True)
plt.show()
```

```python
c1 = df22.drop(['RANK','Whisker-high','Whisker-low',"Year"], axis=1, inplace=False)
corr = c1.corr()
plt.figure(figsize = [7,5])
sn.barplot(y = corr['Happiness Score'].index, x = corr['Happiness Score'])
plt.title("Correlation", fontsize= 15,fontweight="bold")
plt.ylabel("Attributes")
plt.xlabel("Happiness score")
```

# Hypothesis Testing & Linear Regression

```python
df22.columns = df22.columns.str.replace(" ","_")
```

```python
import statsmodels.formula.api as smf
model1 = smf.ols(formula = "Happiness_Score ~ GDP_per_capita + Social_support + Healthy_life_expectancy + Freedom + Trust + Generosity", data = df22).fit()
print(model1.summary())
```

```python
model2 = smf.ols(formula = "Happiness_Score ~ Social_support + Healthy_life_expectancy + Freedom ", data = df22).fit()
print(model2.summary())
```
